```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 計算量のオーダーによってどれくらい性能差が出るのかを描画してみる

# 描画範囲
x = np.arange(0.1, 100, 0.1) # (最小値, 最大値, 刻みの単位)

# 計算式
a = np.log2(x) # O(log(n))
b = x # O(n)
c = x * np.log2(x) # O(nlog(n))
d = x ** 2 # O(n^2)
e = x ** 3 # O(n^3)

# 描画したい変数の設定
plt.plot(x, a)
plt.plot(x, b)
plt.plot(x, c)
plt.plot(x, d)
# plt.plot(x, e) # O(n^3)は大きくなりすぎるのでコメントアウトしている

# 描画
plt.show()
```
